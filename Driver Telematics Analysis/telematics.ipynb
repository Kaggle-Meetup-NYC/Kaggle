```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


%pylab inline
```

```python
path = YOUR DIRECTORY HERE
allFiles = glob.glob(path + "/*.csv")

#create a list
dfs = []
for file_ in allFiles:
    #read in each trip file as a dataframe named df
    df = pd.read_csv(os.path.join(path, file_), index_col=None, header=0)
    
    #strip the file name into a variable named file_name
    file_name = os.path.splitext(os.path.basename(file_))[0]
    
    #create new variable named trip in the data frame with the file_name as its value (constant)
    df['trip'] = pd.Series(int(file_name), index=df.index)
    df['speed'] = np.sqrt(((df['x'] - df['x'].shift(1))**2)+((df['y']-df['y'].shift(1))**2))
    df['accel'] = df['speed']-df['speed'].shift(1)
    df['stop'] = df.apply(lambda row: 1 if (row['speed'] == 0) else 0, axis=1)
    
    #append each df dataframe in dfs dataframe
    dfs.append(df)

total_df = pd.concat(dfs)
```
